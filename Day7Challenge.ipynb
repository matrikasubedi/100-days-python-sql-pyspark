1. How to convert the index of a PySpark DataFrame into a column?

```python
df = spark.createDataFrame([
    ("Alice",1),
    ("Bob",2),
    ("Charlie",3)
],["Name","Value"])

display(df)
```

```python
from pyspark.sql.functions import monotonically_increasing_id
df = df.withColumn("index",monotonically_increasing_id())
display(df)
```

 2. How to combine many lists to form a PySpark DataFrame?

```python
# Define your lists
list1 = ["a", "b", "c", "d"]
list2 = [1, 2, 3, 4]

list_new = list(zip(list1,list2))

df = spark.createDataFrame(list_new, ["column1","column2"])

display(df)
```

4. How to get the items of list A not present in list B?

```python
list_A = [1, 2, 3, 4, 5]
list_B = [4, 5, 6, 7, 8]

result = list(set(list_A)-set(list_B))

print(result)
```

```python
# using join

df_A = spark.createDataFrame([1,2,3,4,5],["value"])
df_B = spark.createDataFrame([4,5,6,7,8],["value"])

result = df_A.join(df_B,on="value",how="left_anti")
display(result)
```

5. How to get the items not common to both list A and list B?

```python
list_A = [1, 2, 3, 4, 5]
list_B = [4, 5, 6, 7, 8]

result = list((set(list_A) - set(list_B)) | (set(list_B) - set(list_A)))
print(result)
```

```python
# using left_anti

resultA = df_A.join(df_B,on="value",how="left_anti")
resultB = df_B.join(df_A,on="value",how="left_anti")
result = resultA.union(resultB)

display(result)
```

6. How to get the minimum, 25th percentile, median, 75th, and max of a numeric column?

```python
# Create a sample DataFrame
data = [("A", 10), ("B", 20), ("C", 30), ("D", 40), ("E", 50), ("F", 15), ("G", 28), ("H", 54), ("I", 41), ("J", 86)]
df = spark.createDataFrame(data, ["Name", "Age"])

df.show()
```

```python
from pyspark.sql import functions as F

df.select(
    F.min("Age").alias("min"),
    F.percentile_approx("Age", 0.25).alias("p25"),
    F.percentile_approx("Age", 0.5).alias("median"),
    F.percentile_approx("Age", 0.75).alias("p75"),
    F.max("Age").alias("max")
).show()
```
