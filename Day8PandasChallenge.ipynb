```python
#  Q1. Create a DataFrame with columns:
# Product, Price , Sales

import pandas as pd

dict = {
    "Product" : ["Laptop","Mouse","Keyboard","Headphone","Charger","Camera","Multiplug"],
    "Price" : [100000,500,1000,1500,800,25000,700],
    "Sales" : [2,4,6,9,8,3,5]
}

df =pd.DataFrame(dict)

print(df)
```

```python
# Q2. Add New Column

df["Revenue"] = df["Price"] * df["Sales"]

print(df)
```

```python
#Q3. Basic Operations

#Print first 3 rows

df.loc[0:2]
```

```python
# Q4. Add New Record - Add a new product using loc

df.loc[len(df)] = ["Watch",10000,8,10000*8]

print(df)
```

```python
# Q5. Mising Data Handling

df.loc[len(df)] = ["Table",10000,8,None]   # added missing values manually

print(df)
```

```python
print(df.isnull())     # Detect using isnull()
```

```python
print(df.isnull().sum()) # count missing values
```

```python
df.fillna(0,inplace=True)   # fill using fillna(0)

print(df)
```

```python
# Q6. low Sales alert
# Display Product where Sales < 8

df[df["Sales"] < 8]
```
