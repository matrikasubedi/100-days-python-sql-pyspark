1- Display the names of employees who have a salary above the average salary in their department

```python
%sql
SELECT 
  emp_name,
  department_name,
  salary,
  avg_dept_salary
FROM (
  SELECT 
    emp_name,
    department_name,
    salary,
    AVG(salary) OVER (PARTITION BY department_id) as avg_dept_salary
  FROM test.new_schema.employees
  WHERE department_id IS NOT NULL
) subquery
WHERE salary >= avg_dept_salary
ORDER BY department_name, emp_name;
```

3-Write a python function to filter out customers who have made more than 5 purchases in the last 6 months

```python
from datetime import datetime, timedelta

customers = [
    {
        "customer_id": 1,
        "purchases": [
            datetime(2026, 3, 1),
            datetime(2026, 2, 15),
            datetime(2026, 1, 10),
            datetime(2025, 12, 20),
            datetime(2025, 11, 5),
            datetime(2025, 10, 1)
        ]
    }
]
```

```python
from datetime import datetime, timedelta

def filter_customers(customers):
    six_months_ago = datetime.now() - timedelta(days=180)

    filtered = []

    for customer in customers:
        recent_purchases = [
            purchase_date
            for purchase_date in customer["purchases"]
            if purchase_date >= six_months_ago
        ]

        if len(recent_purchases) > 5:
            filtered.append(customer)

    return filtered

print(filter_customers(customers))
```

```python
# Assume a table like purchases(customer_id, purchase_date)

SELECT 
 customer_id,
 COUNT(*) AS purchase_count
FROM purchases
WHERE purchase_date >= CURRENT_DATE - INTERVAL '6 months'
GROUP BY customer_id
HAVING COUNT(*) > 5;
```
