## 📅 Day 6 Challenge — SQL & PySpark

### 🎯 Objective:
Analyze customer purchasing behavior and identify high-value customers across franchises.

---

### 🧩 Challenge Question:

Using the tables provided (`sales_transactions`, `sales_customers`, `sales_franchises`):

1. Calculate the **total spending per customer**.
2. Identify the **top 5 customers** based on total spending.
3. For each of these top customers, also return:
   - Full name (`first_name + last_name`)
   - Email address
   - Total number of transactions
   - The **franchise name** where they spent the most money
4. Sort the final result by total spending in descending order.

---

### ⚙️ Requirements:
- Use SQL (and optionally PySpark DataFrame API).
- Handle ties appropriately when ranking top customers.
- Ensure joins are optimized and readable.
- Bonus: Show how you would cache or optimize this query in PySpark.

---

### 🧠 Skills Tested:
- Aggregations (`SUM`, `COUNT`)
- Window functions (for ranking)
- Joins across multiple tables
- Subqueries or CTEs
- PySpark optimization techniques

---

````python
%sql
with cte as (select 
customerID,
franchiseID,
count(transactionID) as count_transaction,
sum(totalPrice) as total_spend
from samples.bakehouse.sales_transactions
group by 1,2)

select 
concat(c.first_name,' ',c.last_name) as Full_name,
c.email_address,
ct.count_transaction,
f.name,
ct.total_spend
from cte ct
join samples.bakehouse.sales_customers c on ct.customerID = c.customerID
join samples.bakehouse.sales_franchises f on ct.franchiseID = f.franchiseID order by total_spend desc limit 5;
````

````python
df_transactions = spark.table("samples.bakehouse.sales_transactions")
df_customers = spark.table("samples.bakehouse.sales_customers")
df_franchises = spark.table("samples.bakehouse.sales_franchises")
````

````python
from pyspark.sql import functions as F
df_agg = df_transactions.groupby("customerID","franchiseID")\
    .agg(F.count(F.col("transactionID")).alias("count_transaction"),(F.sum(F.col("totalPrice")).alias("total_spend")))


````

````python
df_join = df_agg.join(df_customers, on="customerID", how="inner")
df_all = df_join.join(df_franchises,on='franchiseID',how='inner')
df_selected = df_all.select(F.concat(("first_name"),("last_name")).alias("full_name"),"email_address","count_transaction","total_spend","name")

df_selected = df_selected.orderBy(F.col(("total_spend")).desc()).limit(5)
display(df_selected)
````

### 🧩 Problem: "Even Number Counter"

You are given a list of integers. Your task is to:

👉 Count how many numbers in the list are **even**  
👉 Print the total count

---

## 📥 Input
A Python list of integers.

Example:
```python
numbers = [2, 7, 10, 15, 18, 21, 24]

````python
numbers = [ 2,7,10,15,18,21,24]
count = 0
for i in numbers:
    if i % 2 == 0:
        count = count + 1
print(f"Number of even numers in list are : {count}")
````

# 🐍 Day 5 – Python OOP Challenge (Beginner Friendly)

## 🎯 Topic: Classes & Objects

### 🧩 Problem: "Student Report Card"

Create a Python class called **Student** that stores and displays a student's details.

---

## 📥 Requirements

Your class should have:

### Attributes:
- `name` (string)
- `marks` (list of integers)

### Methods:
1. `average()` → returns the average of marks  
2. `result()` → returns:
   - `"Pass"` if average marks ≥ 40
   - `"Fail"` if average marks < 40

---

## 📤 Output Example

If input is:
```python
s = Student("Asha", [50, 60, 70])

````python
class student:

    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    
    def average(self):
        avg = sum(self.marks)/(len(self.marks))
        return avg

    def result(self):
        if self.average() >= 40:
            print("pass")
        else:
            print("fail")

s = student("Asha", [50, 60, 70])

print(s.average())
s.result()
````
