## 🧠 SQL
Find the **top 3 states** with the highest number of customers.

**Conditions:**
- `country = 'United States'`
- `email_address IS NOT NULL`

**Output:**
- `state`
- `customer_count`

**Bonus:** Handle ties for 3rd place.

## ⚡ PySpark

Using DataFrame `df`:

1. Filter:
   - `country == "United States"`
   - `email_address IS NOT NULL`
2. Group by `state` and count
3. Get **top 3 states by customer count**

**Advanced:** Use `dense_rank()` and return `rank <= 3`

````python
%sql
SELECT distinct country FROM samples.bakehouse.sales_customers limit 10;
````

````python
%sql
with cte as (SELECT
state,
count(customerID) as customer_count
FROM samples.bakehouse.sales_customers where country = 'USA' and email_address is not null group by state),

cte2 as (SELECT
*,
DENSE_RANK() over (order by customer_count desc) as rnk
from cte)

select state,customer_count from cte2 where rnk <=3;

````

````python
df = spark.table("samples.bakehouse.sales_customers")
df.show(10)
````

````python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Reload the table to ensure we have fresh data
df = spark.table("samples.bakehouse.sales_customers")

filtered_df = df.filter(
    (F.col("country") == "USA") &
    (F.col("email_address").isNotNull())
)

agg_df = filtered_df.groupBy("state") \
    .agg(F.count("customerID").alias("customer_count"))

win_spec = Window.orderBy(F.col("customer_count").desc())

ranked_df = agg_df.withColumn(
    "rank",
    F.dense_rank().over(win_spec)
)

result_df = ranked_df.filter(F.col("rank") <= 3)

display(result_df)
````
