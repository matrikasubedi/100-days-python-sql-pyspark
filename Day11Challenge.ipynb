## Questions and Solutions

Question 1: Departments with more than 3 employees

```python
%sql
SELECT 
  d.dept_name,
  count(e.emp_id) as total_emp_per_department
FROM departments d
JOIN employees e on d.dept_id = e.dept_id
GROUP BY dept_name 
HAVING count(e.emp_id) >3;
```

Question 2: Find departments with no employees (JOIN)

```python
%sql
SELECT 
  d.dept_name
FROM departments d
LEFT JOIN employees e on d.dept_id = e.dept_id
WHERE e.dept_id IS NULL;
```

Question 3: Find employees without department (JOIN)

```python
%sql
SELECT 
  e.name
FROM employees e
LEFT JOIN departments d on e.dept_id = d.dept_id
WHERE e.dept_id IS NULL;
```

Question 4: Find second highest salary

```python
%sql
WITH rank_table AS (
    SELECT
      salary,
      dense_rank() OVER (ORDER BY salary desc) as rnk
    FROM employees)
SELECT
  salary as second_highest_salary
FROM rank_table
WHERE rnk = 2;
```

Question 5: Count employees hired per year

```python
%sql
SELECT
  YEAR(hire_date) AS hire_year,
  COUNT(*) AS total_employees
FROM employees
GROUP BY YEAR(hire_date)
ORDER BY hire_year;
```

Question 6: Running total of salary (Window Function)

```python
%sql
SELECT 
    emp_id,
    name,
    salary,
    SUM(salary) OVER (ORDER BY emp_id) AS running_total
FROM employees;
```
