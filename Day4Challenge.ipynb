You are given a dataset/table employees with the following columns:

emp_id (int)
emp_name (string)
dept_id (int)
salary (int)

👉 Task:

For each department, find the top 3 employees with the highest salary.

````python
%sql
CREATE TABLE test.new_schema.employees (
    emp_id int,
    emp_name string,
    dept_id int,
    salary int
);
````

````python
%sql
INSERT INTO test.new_schema.employees VALUES
(1, 'Alice', 10, 70000),
(2, 'Bob', 10, 80000),
(3, 'Charlie', 10, 80000),
(4, 'David', 10, 60000),
(5, 'Eve', 20, 90000),
(6, 'Frank', 20, 85000),
(7, 'Grace', 20, 90000),
(8, 'Heidi', 20, 75000),
(9, 'Ivan', 30, 50000),
(10, 'Judy', 30, 55000),
(11, 'Ken', 30, 55000),
(12, 'Leo', 30, 45000);
````

````python
%sql
drop table test.new_schema.employees;
````

````python
%sql 
select * from test.new_schema.employees;
````

````python
%sql 
WITH CTE as (
SELECT
emp_id,
emp_name,
dept_id,
salary,
row_number() over (partition by dept_id order by salary desc) as rnk
from test.new_schema.employees)

SELECT 
emp_name,
dept_id,
salary
from cte where rnk <=3;
````

````python
df = spark.table("test.new_schema.employees")
display(df)
````

````python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

window_spec = Window.partitionBy("dept_id").orderBy(F.col("salary").desc())

df = df.withColumn("rnk", F.row_number().over(window_spec))


df = df.select("emp_name","dept_id","salary").filter(F.col("rnk")<4)

display(df)
````

## Problem: Find First Non-Repeating Character

You are given a string `s`.

### Task:
Write a Python function to find the **first non-repeating character** in the string.  
If all characters repeat, return `None`.

---

### Example:

```python
# Input:  
"aabbcdeff"

# Output:
"c"

````python
str = input("Enter a string")
freq  = {}
for ch in str:
    freq[ch] = freq.get(ch,0) + 1
print(freq)
````

````python
for i in freq:
    if freq[i] == 1:
        print(i)
else:
    None
    
````
