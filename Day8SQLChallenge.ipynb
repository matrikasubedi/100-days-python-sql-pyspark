Scenario-Based SQL Interview Questions
 
1- Find Duplicate Records in a table(amazon)\
2- Retrieve the second highest salary from Employee table\
3- Find Employees without Department (uber)\
4- Calculate the Total Revenue per product (PayPal)\
5- Get the top 3 highest-paid Employees (Google)

Question number 1,2,3,5.

```python
%sql

CREATE TABLE test.new_schema.employees (
    emp_id INT,
    emp_name STRING,
    department_id INT,
    department_name STRING,
    salary DECIMAL(10,2)
);
```

```python
%sql
-- Employees
INSERT INTO test.new_schema.employees VALUES
(1, 'Alice', 10, 'HR', 5000),
(2, 'Bob', 20, 'IT', 7000),
(3, 'Charlie', NULL, NULL, 6000),
(4, 'David', 20, 'IT', 7000),
(5, 'Eve', 30, 'Finance', 9000),
(2, 'Bob', 20, 'IT', 7000); -- duplicate
```

```python
%sql
-- Q1. Find duplicate recods
with cte as (
  SELECT
    *,
    row_number() over (partition by emp_id order by emp_id) as rnk
  FROM test.new_schema.employees)
SELECT * from cte where rnk > 1;
```

```python
%sql

select emp_id , count(*)
from test.new_schema.employees
group by emp_id having count(*) >1;
```

```python
%sql
with cte as (
  SELECT
    *,
    dense_rank() over (order by salary desc) as rnk
  FROM test.new_schema.employees)
SELECT * from cte where rnk = 2;
```

```python
%sql

select * 
from test.new_schema.employees
where department_id is null;
```

```python
%sql
with cte as (
  SELECT
    *,
    dense_rank() over (order by salary desc) as rnk
  FROM test.new_schema.employees)
SELECT * from cte where rnk  <= 3;
```

```python
%sql
CREATE TABLE test.new_schema.products (
    product_id INT,
    product_name STRING,
    revenue DECIMAL(10,2)
);
```

```python
%sql
-- Products
INSERT INTO test.new_schema.products VALUES
(101, 'ProductA', 200),
(102, 'ProductB', 300),
(101, 'ProductA', 150),
(103, 'ProductC', 400),
(102, 'ProductB', 500);
```

```python
%sql
Select product_id ,
product_name,
sum(revenue) as total_revenue
from test.new_schema.products group by product_id,product_name;
```
